## Attriton with Monthly Income
* H0 is that attrition is covariant with Monthly Income
* H1 is that attritionn is not covariant with Monthly Income

```
stats, p = pearsonr(dataset1.Attrition, dataset1.MonthlyIncome)
stats, p
```

p value is less than 0.05 so we can accept the H1
